```python
from zipfile import ZipFile
```

```python
with ZipFile('mmt_gtfs.zip') as zf:
    for info in zf.infolist():
        print(info)
```

```python
with ZipFile('mmt_gtfs.zip') as zf:
    for info in zf.infolist():
        orig_mb = info.file_size / (1024**2) # there are 1024**2 bytes in a MB
        ratio = info.file_size / info.compress_size
        s = "file {name:s}, {mb:.3f} MB (uncompressed), {ratio:.1f} compression ratio"
        print(s.format(name=info.filename, mb=orig_mb, ratio=ratio))
```

```python
with ZipFile('mmt_gtfs.zip') as zf:
    orig_mb = 0
    compress_mb = 0
    for info in zf.infolist():
        orig_mb = orig_mb + info.file_size / (1024**2) # there are 1024**2 bytes in a MB
        compress_mb = compress_mb + info.compress_size / (1024**2)
    ratio = orig_mb / compress_mb
    s = "file {name:s}, {mb:.3f} MB (uncompressed), {ratio:.1f} compression ratio"
    print(s.format(name='mmt_gtfs.zip', mb=orig_mb, ratio=ratio))
```
