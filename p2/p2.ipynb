```python
import matplotlib.pyplot as plt
def get_ax():
    ax = plt.subplot()
    ax.set_xticks([-8, -6, -4, -2, 0, 2, 4, 6, 8])
    ax.set_yticks([-8, -6, -4, -2, 0, 2, 4, 6, 8])
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    #https://blog.csdn.net/sinat_34328764/article/details/80246139
    ax.xaxis.set_ticks_position('bottom')   
    ax.yaxis.set_ticks_position('left')
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data', 0))
    return ax
```

```python
bd.BST_stops.sec_line()
line_data = bd.BST_stops.line_df
line_data.head(20)
```

```python
ax = get_ax()
df_stops_ava_wheel = bd.df_stops_ava[bd.df_stops_ava['wheelchair_boarding'] == 1]
df_stops_ava_wheeln = bd.df_stops_ava[bd.df_stops_ava['wheelchair_boarding'] == 0]
df_stops_ava_wheel.plot.scatter('x','y', ax = ax, color = "red", figsize=(10, 10))
df_stops_ava_wheeln.plot.scatter('x','y', ax = ax, color = "0.7")
for i in range(len(line_data)):
    position = line_data.iloc[i]['position']
    lim1 = line_data.iloc[i]['lim'][0]
    lim2 = line_data.iloc[i]['lim'][1]
    linewidth = line_data.iloc[i]['level']
    if line_data.iloc[i]['direction'] == 'x':
        ax.plot((position, position), (lim1, lim2), lw=10/linewidth, color="yellow")
    else:
        ax.plot((lim1, lim2), (position, position), lw=10/linewidth, color="yellow")
```
